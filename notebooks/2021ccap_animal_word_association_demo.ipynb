どうぶつの名前を想起する課題のモデル化の試み。
もちろん，第一歩のたたき台モデルです。

以下にモデルの概要です。

1. 多くの動物単語ベクトルで張られる空間へ単語ベクトルを射影し，
2. その射影空間内で類似したベクトルの想起 model
3. 各ベクトルの類似度をソフトマックス関数を用いて確率密度に変換
4. 変換した確率密度を元にサンプリングする
